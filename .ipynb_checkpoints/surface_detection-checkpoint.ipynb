```python
import numpy as np
import matplotlib.pyplot as plt
import ipyvolume as ipv
from scipy.ndimage import gaussian_filter
```

```python
DATA_PATH = 'C:/Gel_Sheet_Data/'
MOVIE_PATH = DATA_PATH +'movie60/'
gel = np.load(MOVIE_PATH +'np/gel.npy', mmap_mode='r')

```

```python
ipv.figure()
ipv.volshow(gel[10].astype(np.float64))
ipv.show()
```

```python
import numpy as np
import ipyvolume as ipv
V = np.zeros((128,128,128)) # our 3d array
# outer box
V[30:-30,30:-30,30:-30] = 0.75
V[35:-35,35:-35,35:-35] = 0.0
# inner box
V[50:-50,50:-50,50:-50] = 0.25
V[55:-55,55:-55,55:-55] = 0.0

ipv.figure()
ipv.volshow(V, level=[0.25, 0.75], opacity=0.03, level_width=0.1, data_min=0, data_max=1)
ipv.view(-30, 40)
ipv.show()
```

```python
plt.imshow(np.ones((5,5)))
plt.show()
```

```python
time_point = gel[35]
#plt.figure()
#for i  in range(0,512,50):
#    plt.plot(time_point[256,i,:])
bg = gaussian_filter(np.mean(time_point[:,:,:],axis = 2),25)
# plt.imshow(bg); plt.colorbar()
imstack_bg = np.zeros(time_point.shape)
for i in range(len(gel)):
    imstack_bg[:,:,i] = gaussian_filter(time_point[:,:,i],2)/bg
plt.figure()
for i  in range(0,512,50):
    plt.plot(imstack_bg[i,256,:])
# plt.ylim([-.2,.2])
```

```python
h = np.zeros(gel[0][0].shape)
for i in range(h.shape[0]):
    for j in range(h.shape[1]):
        h[i,j] = np.where(imstack_bg[i,j,:]>1.05)[0][-1] 
```
