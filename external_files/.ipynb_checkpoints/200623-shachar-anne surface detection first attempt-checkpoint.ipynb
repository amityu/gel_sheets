## <span color=red >Load Images

</span>


```python
#get list of files in directory
import os
filelist = os.listdir(MOVIE_PATH + 'tiff')
file_list_numbered = []
for i in range(len(filelist)):
    #split file by '_'
    t = int(filelist[i].split('_')[-2][1:])
    file_list_numbered.append((t,filelist[i]))
    #sort by time
file_list_numbered.sort(key = lambda x: x[0])
for t in range(len(file_list_numbered)):
    print(file_list_numbered[t][0],file_list_numbered[t][1])


```

```python
#load images
gel_list = []
for i in range(len(file_list_numbered)):
    gel_list.append(tifffile.imread(MOVIE_PATH + 'tiff/' + file_list_numbered[i][1]))
    print(i)
```

```python
# get the max z since images where taken with different z
max_z = 0
for image in gel_list:
    print(image.shape)
    z = image.shape[0]
    if z > max_z:
        max_z = z
print(max_z)
```

```python
#set blank areas to nan
gel = np.zeros((len(gel_list), max_z,*gel_list[0].shape[1:]))
gel[gel ==0] = np.nan
for t in range(len(gel)):
    gel[t,:gel_list[t].shape[0],:,:] = gel_list[t]
```

```python
np.save(MOVIE_PATH + 'np/gel.npy',gel)
```

```python
plt.figure()
for i  in range(0,512,50):
    plt.plot(im_stack[:,256,i])

plt.title(' Raw data '  + movie + " t = %d intensity profile in z "%t)
```

```python
bg_below_plate = gaussian_filter(np.mean(im_stack[0:5,:,:],axis = 0),25)
plt.imshow(bg_below_plate); plt.colorbar(); plt.title('Illumination below plate')
```

```python
mean_unevened_list = []
std_unevened_list = []
mean_evened_list = []
std_evened_list = []
t_gap = 3
for t in range(len(gel)-t_gap):
    moving_stack = gel[t]
    moving_bg_below_plate = gaussian_filter(np.mean(moving_stack[0:5,:,:],axis = 0),25)
    next_moving_bg = gel[t+t_gap][0:5,:,:]

    mean_unevened,std_unevened = np.mean(next_moving_bg),np.std(next_moving_bg)/np.mean(moving_bg_below_plate)
    mean_unevened_list.append(mean_unevened)
    std_unevened_list.append(std_unevened)

    mean_evened,std_evened = np.mean(next_moving_bg/moving_bg_below_plate),np.std(next_moving_bg/moving_bg_below_plate)
    mean_evened_list.append(mean_evened)
    std_evened_list.append(std_evened)
plt.plot(mean_unevened_list, label = 'mean_t')
#plt.plot(mean_evened_list, label = 'mean_evened t+gap', linestyle = 'dotted')
plt.title('mean of bg below plate normalized by mean of bg below plate of t')
plt.legend()
plt.show()
plt.plot(std_unevened_list, label = 'std_unevened')

plt.plot(std_evened_list, label = 'std_evened')
plt.legend()
plt.title('mean and std of bg below plate')
plt.show()
```

```python
mean_list = []
for t in range(len(gel)):
    mean_list.append(np.mean(gel[t,0:5,:,:]))

plt.plot(mean_list, label = 'mean_bg', linestyle = '-')
plt.title('mean of bg below plate')
plt.legend()
plt.xlabel('t')
plt.ylabel('mean')
plt.show()
```
