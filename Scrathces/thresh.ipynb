```python
#import np,gaussian
import numpy as np
from skimage.filters import gaussian
import matplotlib.pyplot as plt
```

```python
frequency = np.zeros(100)
frequency[49:51] = 100
frequency = gaussian(frequency, 10)
frequency = frequency / np.max(frequency)
plt.plot(frequency)
plt.xlabel('Intensity')
plt.ylabel('Frequency')

plt.show()
```
